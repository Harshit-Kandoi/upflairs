# The matplotlib library in Python provides a wide range of functions for creating static, animated, and interactive visualizations.

# Here are some of the key functions in matplotlib, specifically focusing on the pyplot module, which is commonly used for creating plots:
# Basic Plotting

plt.plot(): Plots 2D line plots.

plt.scatter(): Creates a scatter plot.

plt.bar(): Creates a bar plot.

plt.barh(): Creates a horizontal bar plot.

plt.hist(): Plots a histogram.

plt.boxplot(): Creates a box plot.

plt.pie(): Creates a pie chart.

# Customization

plt.title(): Adds a title to the plot.

plt.xlabel(): Adds a label to the x-axis.

plt.ylabel(): Adds a label to the y-axis.

plt.xlim(): Sets the limits for the x-axis.

plt.ylim(): Sets the limits for the y-axis.

plt.xticks(): Sets the ticks on the x-axis.

plt.yticks(): Sets the ticks on the y-axis.

plt.legend(): Adds a legend to the plot.

plt.grid(): Adds a grid to the plot.

# Subplots

plt.subplot(): Creates subplots in a single figure.

plt.subplots(): A more flexible way to create subplots.

plt.subplot2grid(): Creates subplots with a grid layout.

# Figure and Axes

plt.figure(): Creates a new figure.

plt.gcf(): Gets the current figure.

plt.gca(): Gets the current axes.

plt.clf(): Clears the current figure.

plt.cla(): Clears the current axes.

# Image and Contour Plots

plt.imshow(): Displays an image.

plt.contour(): Creates a contour plot.

plt.contourf(): Creates a filled contour plot.

# Saving Plots

plt.savefig(): Saves the current figure to a file.

Displaying Plots

plt.show(): Displays the plot.

# Line plot graph using matplotlib

```python
random_ls =[5,6,8,6,8]   #list of number we want to plot
plt.plot(random_ls)      #creates a line plot of the numbers
plt.plot(random_ls,marker='*',ms=15,mfc='red')   # marker = design for plotting 
# we can use different different kind of marker for plotting
# ms = marker size     # mfc = marker face color
```

```python
axis_1 = [3,5,7,4,6,9]   #creating a list named 'axis_1'
axis_2 = [4,6,8,9,2,3]   #creating a list named 'axis_2'
# Plotting 'axis_2' against 'axis_1' with star markers
plt.plot(axis_2,axis_1,marker='*',ms=15,mfc='red')
plt.title("lineplot")  # Adding a title to the plot
plt.xlabel('axis_1')   # Labeling the x-axis
plt.ylabel('axis_2')   # Labeling the y-axis
plt.show()             # Displaying the plot
```

```python
axis_1 = [3,5,7,4,6,9]
axis_2 = [4,6,8,9,2,3]
# Plotting 'axis_1' against 'axis_2' with star markers
plt.plot(axis_1,axis_2,marker='*',ms=20,mfc='red',color = 'k', lw= 4) 
# k = black, lw = linewidth, ms = size of marker
```

Some examples of marker you can use for plotting.

```python
plt.figure(figsize = (10,6))
plt.bar(brand, no_of_count, color='dodgerblue')
plt.xticks(rotation=60)
plt.xlabel('Brands',fontdict = {'fontsize':20})
plt.ylabel('no of bike avaliable')
plt.title('bike in showroom',fontdict={'fontsize':20,'color':'blue'})
#plt.yticks
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.bar(brand, no_of_count, color='dodgerblue',ec='k', ls = 'dashed')  #ls - line style, dotted,dashed,solid
plt.xticks(rotation=60)
plt.xlabel('Brands',fontdict = {'fontsize':20})
plt.ylabel('no of bike avaliable')
plt.title('bike in showroom',fontdict={'fontsize':20,'color':'blue'})
#plt.yticks
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.bar(brand, no_of_count, color='green',ec='k', ls = 'dashed',lw = 1.5,width = 1)  #ls - line style
plt.xticks(rotation=60)
plt.xlabel('Brands',fontdict = {'fontsize':20})
plt.ylabel('no of bike avaliable')
plt.title('bike in showroom',fontdict={'fontsize':20,'color':'blue'})
#plt.yticks
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.bar(brand, no_of_count, color='green',ec='k', ls = 'dashed',lw = 1.5,width = 1)  #ls - line style
plt.xticks(rotation=60)
plt.xlabel('Brands',fontdict = {'fontsize':20})
plt.ylabel('no of bike avaliable')
plt.title('bike in showroom',fontdict={'fontsize':20,'color':'blue'})
plt.savefig('bike_brands')
#plt.yticks
plt.show()
```

pie plot()

```python
population = [700,560,450,510,300]
country = ['India','china','japan','bangladesh','nepal']
plt.pie(population,labels = country)
plt.show()
```

```python
population = [700,560,450,510,300]
country = ['India','china','japan','bangladesh','nepal']
plt.pie(population,labels = country, autopct = '%1.2f%%',wedgeprops={'ec':'k'})
plt.title('Country population distribution in pie plot')
plt.show()
```

```python
population = [700,560,450,510,300]
country = ['India','china','japan','bangladesh','nepal']
exp = [0,0,0,0,0.2] 
plt.pie(population,labels = country, autopct = '%1.2f%%',wedgeprops={'ec':'k'},explode=exp)
plt.title('Country population distribution in pie plot')
plt.show()
```

```python
population = [700,560,450,510,300]
country = ['India','china','japan','bangladesh','nepal']
exp = [0,0.3,0,0,0.2] 
plt.pie(population,labels = country, autopct = '%1.2f%%',wedgeprops={'ec':'k'},explode=exp)
plt.title('Country population distribution in pie plot')
plt.show()
```

```python
population = [700,560,450,510,300]
country = ['India','china','japan','bangladesh','nepal']
exp = [0,0.3,0,0,0.2] 
plt.pie(population,labels = country, autopct = '%1.2f%%',wedgeprops={'ec':'k','ls':'dashed'},explode=exp,shadow = True)
plt.title('Country population distribution in pie plot')
plt.show()
```
